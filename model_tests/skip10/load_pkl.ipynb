```python
import pickle
import numpy as np

# Load the pickle dataset
data_path = "test_nov22_no_trace.pkl"
with open(data_path, 'rb') as f:
    data = pickle.load(f)

# # Extract X and Y
# X = np.array(data.iloc[:-10])  # excluding the last 10 rows
# Y = np.zeros_like(X)              # initialize Y with zeros

# # Define the column index
# column_index = 53

# # Populate Y with the specified conditions
# for i in range(len(Y)):
#     Y[i, column_index] = np.sum(data.iloc[i:i+10, column_index])

# # Verify the shapes of X and Y
# print("Shape of X:", X.shape)
# print("Shape of Y:", Y.shape)
```

```python
X = np.array(data)
total_diff = 0
GAP = 20
column_index = 53
for i in range(len(X) - GAP): #len(X) - 20):
    curr_diff = abs(X[i + GAP, column_index] - X[i, column_index])
    total_diff += curr_diff



print("Average gap skipping {} steps: {}".format(GAP, total_diff / (len(X) - GAP)))
print("Min value in dataset", np.min(X))
print("Max value in dataset", np.max(X))
```

```python
import matplotlib.pyplot as plt

our_data = np.array(data)[:,column_index]

# Define the downsampling factor (e.g., taking every 10th data point)
downsample_factor = 10

# Downsample the data
downsampled_data = our_data[::downsample_factor]

# Create an array of corresponding indices for the downsampled data
downsampled_indices = np.arange(0, len(our_data), downsample_factor)

# Plot the downsampled time series
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.plot(downsampled_indices, downsampled_data, label=f'Time Series Data (Downsampled by {downsample_factor})')
plt.title('Downsampled Time Series Visualization')
plt.xlabel('Time')
plt.ylabel('PIT300')
plt.show()
```
