```python
import math
import numpy as np
import pandas as pd

L = 600

y = []
for i in np.arange(0.0, L, 0.3):
    print(i)
    y.append(math.sin(i))
```

```python
df = pd.DataFrame({
    'Xt': y[1:],  # Xt values, starting from the second value
    'Xt-1': y[:-1]  # Xt-1 values, excluding the last value
})

split_index = int(0.8 * len(df))

# Split the DataFrame into train and test sets
train_df = df.iloc[:split_index]
test_df = df.iloc[split_index:]
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
# Save the DataFrame to a pickle fil
train_df.to_pickle("train_sine.pkl")
test_df.to_pickle("test_sine.pkl")
```
