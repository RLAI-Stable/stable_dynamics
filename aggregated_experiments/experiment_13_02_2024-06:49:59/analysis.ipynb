# Load files

```python
import pandas as pd
pd.options.display.float_format = '{:.8}'.format


# Read CSV files into pandas DataFrames
model_eval_df = pd.read_csv('model_eval.csv')
model_epochs_df = pd.read_csv('model_epochs.csv')
```

# Filter epochs DF to get only the test loss

```python
model_epochs_df = model_epochs_df[model_epochs_df['train_or_test'] == 'TEST']
display(model_epochs_df)
```

# Get only rows which contain the last training epoch

```python
min_loss_idx_per_uuid = model_epochs_df.groupby('uuid')['epoch'].idxmax()
min_loss_per_uuid_df = model_epochs_df.loc[min_loss_idx_per_uuid]
min_loss_per_uuid_df
```

# Is the minimum test loss correlated to the rollout performance?
It doesn't seem to have such a strong correlation, but that is also due to the fact that the models have very similar performance on the test dataset

```python
# Merge DataFrames on 'uuid' column
merged_df = pd.merge(model_eval_df, min_loss_per_uuid_df, on=['uuid'])

# Perform further analysis or visualization
# For example, calculate correlation between rollout_error and min test loss
correlation = merged_df['rollout_error'].corr(merged_df['loss'])
print("Correlation between rollout_error and min test loss:", correlation)
```

```python
# Display the merged DataFrame
merged_df.sort_values(by='rollout_error')
```

# Print the directory paths of the best performing models

```python
best_models_df = merged_df.sort_values(by='rollout_error', ascending=True).head(5)[["model_parameters_x", "uuid"]]
```

```python
best_models_df['path'] = best_models_df.apply(lambda row: row["model_parameters_x"] + "/" + row["uuid"], axis=1)
```
